```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = "C://Users//vibhu//OneDrive//Desktop//IIITDelhi//ComputerVision//sex//Panorama-master//Panorama-master//data//building"

image1=cv2.imread(path+"//building1.jpg")
image2=cv2.imread(path+"//building2.jpg")

image3=cv2.imread(path+"//building3.jpg")

image4=cv2.imread(path+"//building4.jpg")

image5=cv2.imread(path+"//building5.jpg")


image1=cv2.resize(image1, (0,0), fx=0.5, fy=0.5)
image2=cv2.resize(image2, (0,1), fx=0.5, fy=0.5)
image3=cv2.resize(image3, (0,2), fx=0.5, fy=0.5)

image4=cv2.resize(image4, (0,3), fx=0.5, fy=0.5)
image5=cv2.resize(image5, (0,4), fx=0.5, fy=0.5)


image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

image5 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)




plt.subplot(121),plt.imshow(image1)
plt.title('Image 1'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image2)
plt.title('Image 2'), plt.xticks([]), plt.yticks([])
plt.subplot(121),plt.imshow(image3)
plt.title('Image 3'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image4)
plt.title('Image 4'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image5)
plt.title('Image 5'), plt.xticks([]), plt.yticks([])
plt.show()
```
